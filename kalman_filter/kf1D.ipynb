```python
from math import *
import matplotlib.pyplot as plt
import numpy as np
import sys
import numpy as geek
```

```python
measurements = [3.,4.,6.,7.,8.]
motion = [1., 2., 1., 1., 1.]
measurement_sigma = 1.  #measurement uncertainty
motion_sigma = 2.	    #motion uncertainty
mu = 0				    #initial position estimate (and mu later represents the position)
position_sigma = 1. #position certainty
```

```python
def f(mu, sigma2, x):
  coefficient = 1.0 / sqrt(2.0 * pi *sigma2)
  exponential = exp(-0.5 * (x-mu) ** 2 / sigma2)
  return coefficient * exponential

#blue update
#orange prediction
def plot(mu_update,sig_update,mu_predict,sig_predict):

  # define a range of x values
  x_axis = np.arange(-5, 15, 0.1)
  # create a corresponding list of gaussian values
  g_update = []
  for x in x_axis:
    g_update.append(f(mu_update, sig_update, x))

  g_predict = []
  for x in x_axis:
    g_predict.append(f(mu_predict, sig_predict, x))

  # plot the result
  plt.plot(x_axis, g_update)
  plt.plot(x_axis, g_predict)
  plt.show()

#TODO #def update
def update(mean_position,var_position,mean_measurement,var_measurement):
  new_mean = (var_measurement*mean_position+var_position*mean_measurement)/(var_measurement+var_position)
  new_var  = 1/((1/var_measurement)+(1/var_position))
  return [new_mean,new_var]
#TODO #def prediction

def prediction(mean_position,var_position,mean_motion,var_motion):
  new_mean = mean_position+mean_motion
  new_var  = var_position+var_motion
  return [new_mean,new_var]
```

```python

for n in range(len(measurements)):
	[mu,position_sigma] = update(mu, position_sigma, measurements[n],measurement_sigma)
	print ('update: ',[mu,position_sigma])

	#just to plot
	mu_u=mu
	position_sigma_u=position_sigma

	[mu,position_sigma] = prediction(mu, position_sigma, motion[n],motion_sigma)
	print ('predict: ',[mu,position_sigma])

	plot(mu_u,position_sigma_u,mu,position_sigma)

#measurements = [3.,4.,6.,7.,8.]
#motion = [1., 2., 1., 1., 1.]
#blue update
#orange prediction
```
